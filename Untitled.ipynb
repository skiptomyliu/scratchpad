```python
import tensorflow as tf
```

```python
import matplotlib.pyplot as plt
from scipy.misc import imread, imresize

%matplotlib inline
# plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
# plt.rcParams['image.interpolation'] = 'nearest'
# plt.rcParams['image.cmap'] = 'gray'
```

```python
filename = "pier.JPG"
raw_image_data = imread(filename)
print raw_image_data.shape

H,W,_ = raw_image_data.shape

image = tf.placeholder("uint8", [None, None, 3])
index_p = tf.placeholder("int32")
# blah = tf.Variable(0, name='index_p')
sliced = tf.slice(image, [0, 0, 0], tf.pack([H/1,W/(2+2*index_p),-1]))

results = []
with tf.Session() as session:
    for i in xrange(2):
        print i
        result = session.run(sliced, feed_dict={image:raw_image_data, index_p:i})
        print result.shape
        results.append(result)
    
for result in results:
    plt.imshow(result)
    plt.show()
```
